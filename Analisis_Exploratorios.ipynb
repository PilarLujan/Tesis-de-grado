```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D

# Función para crear gráficos con leyenda de códigos en el texto
def plot_with_legend(data, category_column, label_column, title):
    # Contar las frecuencias de cada categoría
    count_data = data.groupby(category_column)[label_column].apply(list)

    # Crear el gráfico
    plt.figure(figsize=(20, 8))
    ax = sns.countplot(data=data, x=category_column, palette="viridis", order=count_data.index)

    # Crear un conjunto para almacenar los códigos que ya han sido añadidos a la leyenda
    added_codes = set()
    legend_elements = []

    for cat, codes in count_data.items():
        # Iterar sobre los códigos en la lista
        for code in codes:
            # Solo añadir códigos únicos a la leyenda
            if code not in added_codes:
                added_codes.add(code)  # Agregar a los códigos añadidos
                # Añadir un elemento a la leyenda con el color correspondiente
                legend_elements.append(Line2D([0], [0], marker='o', color='w', label=f"{cat}: {code}",
                                               markerfacecolor=ax.patches[count_data.index.get_loc(cat)].get_facecolor(), markersize=8))

    # Añadir la leyenda dentro del gráfico
    plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1, 1), fontsize=7, title="Leyenda")

    # Configuración del gráfico
    plt.title(title)
    plt.xlabel(category_column)
    plt.ylabel("Frecuencia")
    plt.xticks(rotation=45)
    plt.tight_layout()  # Ajustar el área del gráfico
    plt.show()

# Gráfico del Tiempo de Crecimiento con leyenda de códigos
plot_with_legend(ais, 'Tipo de crecimiento', 'Código de la placa', 'Distribución del Tiempo de Crecimiento con Códigos de Placas')

# Gráfico de la Forma de la Colonia con leyenda de códigos
plot_with_legend(ais, 'Forma de la colonia', 'Código de la placa', 'Distribución de la Forma de la Colonia con Códigos de Placas')

# Gráfico de la Elevación de la Colonia con leyenda de códigos
plot_with_legend(ais, 'Elevación de la colonia', 'Código de la placa', 'Distribución de la Elevación de la Colonia con Códigos de Placas')

# Gráfico del Tamaño de la colonia con leyenda de códigos
plot_with_legend(ais, 'Tamaño de la colonia', 'Código de la placa', 'Distribución del Tamaño de la Colonia con Códigos de Placas')

# Gráfico de la Colonia Translúcida con leyenda de códigos
plot_with_legend(ais, 'Colonia traslucida', 'Código de la placa', 'Distribución de Colonias Translúcidas con Códigos de Placas')

# Gráfico de la Colonia Translúcida con leyenda de códigos
plot_with_legend(ais, 'Color de la colonia', 'Código de la placa', 'Distribución del Color de la colonia con Códigos de Placas')

# Gráfico de la Colonia Gomosa con leyenda de códigos
plot_with_legend(ais, 'Colonia gomosa', 'Código de la placa', 'Distribución de Colonias Gomosas con Códigos de Placas')

# Gráfico de la Colonia Opaca con leyenda de códigos
plot_with_legend(ais, 'Colonia opaca', 'Código de la placa', 'Distribución de Colonias Opacas con Códigos de Placas')

# Gráfico de la Colonia Opaca con leyenda de códigos
plot_with_legend(ais, 'Colonia', 'Código de la placa', 'Distribución de Colonias Opacas con Códigos de Placas')
```
